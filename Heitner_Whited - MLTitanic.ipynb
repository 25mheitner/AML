```python
df['FareCode'] = pd.cut(df['Fare'], bins=10, labels = False)
df['AgeCode'] = pd.cut(df['Age'], bins=10, labels = False)
```

```python
Binary = { 'C':0, 'E':1, 'G':2, 'D':3, 'A':4, 'B':5, 'F':6, 'T':7}
df['CabinCode'] = df['CabinFirstLetter'].map(Binary)
```

```python
# Extract last names by splitting at the comma
df["Last_Name"] = df["Name"].str.split(",").str[0]

# Count occurrences of each last name
last_name_counts = df["Last_Name"].value_counts()

# Create the new column: 1 if last name appears more than once, else 0
df["Shared_Last_Name"] = df["Last_Name"].map(lambda x: 1 if last_name_counts[x] > 1 else 0)

# Drop the helper column if not needed
df.drop(columns=["Last_Name"], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

features = ['Shared_Last_Name', 'CabinCode', 'Fare',  'EmbarkedCode',
       'SexCode', 'FareCode', 'AgeCode', 'SibSp',
       'Parch',"Pclass"]

X = df[features]
y = df['Survived']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)


# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf*100)


```
